```python
delta = 0.025
x = np.arange(-3,3,delta)
y = np.arange(-2,2,delta)

xx, yy = np.meshgrid(x,y)
```

```python
Z1 = np.exp(-xx**2 - yy**2)
Z2 = np.exp(-(xx-1)**2 - (yy-1)**2)
Z = (Z1 - Z2) * 2
```

```python
fig, ax = plt.subplots()
ax.contourf(xx,yy,Z)
ax.set_title('Simple meshgrid example')
ax.set_xlabel('x')
ax.set_ylabel('y')
```
